```python
img = cv2.imread('inputfile/data_table_example1.png')
# im = imorg[0:100, 0:300]
im = img.copy()


imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,110,255,cv2.THRESH_BINARY)
```

```python
# imgray[0:100, 0:200]
```

```python
image1, contours1, hierarchy1 = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
image2, contours2, hierarchy2 = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
for i in contours1:
    perimeter=cv2.arcLength(i,True)
    epsilon1 = 0.05*perimeter
    approx1 = cv2.approxPolyDP(i,epsilon1,True)
    if perimeter > 100:
        img2c = cv2.drawContours(im, i, -1, (255,0,0), 3)
        imgaprox1 = cv2.drawContours(img, [approx1],-1, (255,0,255), 3)
#         print(perimeter)
        list1.append(i)
```
