```python
import cv2
import numpy as np
```

```python
imgslant = cv2.imread('inputfile/rbx.png')
```

```python
imgray = cv2.cvtColor(imgslant,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,127,255,0)
image,contours,hierarchy = cv2.findContours(thresh, 1, 2)

cnt = contours[0]
```

```python
imgc = cv2.drawContours(imgslant,cnt,0,(0,0,255),2)
```

```python
cv2.imshow('img',imgc)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rect = cv2.minAreaRect(cnt)
```
